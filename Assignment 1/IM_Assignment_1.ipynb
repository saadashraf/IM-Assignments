Imports


```python
import math
from matplotlib import pyplot as plt
```

Code for initiating the data

```python
input = [(x + 1) * 0.1 for x in range(70)]
k = 4

log_input = [math.log(x) for x in input]
linear_input = input
quadratic_input = [x ** 2 for x in input]
cubic_input = [x ** 3 for x in input]
polynomial_input = [x ** k for x in input]
```

Plotting

```python
plt.figure(figsize=(16 , 8) , facecolor='white')

plt.plot(input, log_input)
plt.plot(input, linear_input)
plt.plot(input, quadratic_input)
plt.plot(input, cubic_input)
plt.plot(input, polynomial_input)
plt.ylim([0, 400])
plt.xlabel("Input")
plt.ylabel("Complexity")
plt.legend(["Log", "Linear", "Quadratic", "Cubic", "Polynomial"])
```
